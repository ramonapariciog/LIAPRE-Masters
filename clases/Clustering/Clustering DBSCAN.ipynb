# Clustering DBSCAN

```python
%pylab inline
```

```python
import numpy as np
from sklearn import cluster
from sklearn.datasets import make_circles

n_samples = 1000
x, y = make_circles(n_samples=n_samples, factor=0.25,
                    noise=0.05)
scatter(x[:,0],x[:,1],c= 'red', marker = 'o', s=50)
grid()
```

```python
dbscan = cluster.DBSCAN(eps=.2)
y_dbscan = dbscan.fit(x)
y_pred = dbscan.labels_.astype(np.int)
colors = np.array([yy for yy in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)
# scatter(x[:,0], x[:,1], color=colors[y_pred].tolist(), s=10)
scatter(x[:,0], x[:,1], c=dbscan.labels_, s=10)
xticks(())
yticks(())
```

Nota: en vez de usar color usa c y asigna un nmero
