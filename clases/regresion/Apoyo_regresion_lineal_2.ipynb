```python
def aproxima(x, y):
    x = x.reshape(-1,1)
    model.fit(X=x, y=y)
    return model.intercept_, model.coef_
def lineal_model(sigma, N=100):
    ruido = np.random.normal(size=N)
    x = np.linspace(0, 20, N)
    y = (5*x) + 1 + (ruido*sigma)
    # plt.plot(x, y, 'ro')
    return x, y
fig, axes = plt.subplots(1,2)
axes = axes.reshape(-1,)
y0l = []
ml = []
for sigma in range(10, 1000):
    x, y = lineal_model(sigma)
    y0, m = aproxima(x, y)
    y0l.append(y0)
    ml.append(m)
axes[0].plot(y0l)
axes[1].plot(ml)
fig.show()
plt.show()
```

```python
def aproxima(x, y):
    x = x.reshape(-1,1)
    model.fit(X=x, y=y)
    return model.score(x, y)
def lineal_model(sigma, N=100):
    ruido = np.random.normal(size=N)
    x = np.linspace(0, 20, N)
    y = (5*x) + 1 + (ruido*sigma)
    x = x.reshape(-1,1)
    # plt.plot(x, y, 'ro')
    return x, y
fig, axes = plt.subplots(1,2)
axes = axes.reshape(-1,)
scores = []
for sigma in range(10, 1000):
    x, y = lineal_model(sigma)
    score = aproxima(x, y)
    scores.append(score)
axes[0].plot(scores)
fig.show()
plt.show()
```
