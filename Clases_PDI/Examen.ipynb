```python
image = cv2.imread("images/lena_s.png")
```

```python
B, G, R = cv2.split(image)
```

```python
plt.imshow(image[:,:,0], cmap=plt.cm.Blues_r)
```

```python
plt.imshow(R, cmap=plt.cm.Reds_r)
```

```python
import numpy as np
```

```python
cv2.imwrite("lena_cambiada.png", np.stack([R,G,B], axis=2), "PNG")
```

```python
h = np.array([[1,2,1], [0,0,0], [-1,-2,-1]])
h
```

```python
from scipy.signal import convolve2d
```

```python
res = convolve2d(R, h)
```

```python
type(res)
```

```python
res.dtype
```

# Ajuste de rango

```python
cambio1 = res + np.abs(res.min())
imageres = ((cambio1 / cambio1.max())*255).astype(np.uint8)
```

```python
cv2.imwrite("resultado.png", 16, imageres) #, cv2.IMWRITE_PNG_COMPRESSION)
```
