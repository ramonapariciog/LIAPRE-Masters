```python
gray_bill = cv2.cvtColor(billete, cv2.COLOR_RGB2GRAY)
```

```python
gray_bill.shape
```

```python
plt.figure(figsize=(10,10))
plt.imshow(gray_bill, cmap=plt.cm.gray)
plt.axis("off")
plt.show()
```

```python
wt = pywt.wavedec2(gray_bill, "haar", mode="periodization", level=3)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wt[0], cmap=plt.cm.gray)
plt.axis("off")
plt.show()
```

```python
wt = pywt.dwt2(gray_bill, 'haar')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
axes = axes.reshape(-1,)
for ia, (ax, title) in enumerate(zip(axes, ["A", "cH", "cV", "cD"])):
    if ia < 1:
        ax.imshow(wt[0], cmap=plt.cm.gray)
        ax.set_title(title)
        ax.set_axis_off()
    else:
        ax.imshow(wt[1][ia-1], cmap=plt.cm.gray)
        ax.set_title(title)
        ax.set_axis_off()
plt.show()
```
